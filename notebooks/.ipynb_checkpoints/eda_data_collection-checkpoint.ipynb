# üì¶ EDA - Exploration des donn√©es

```python
#Chargement des librairies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Chargement des donn√©es
colis_df = pd.read_csv("../data/synthetic/colis_synthetiques.csv")
meteo_df = pd.read_csv("../data/raw/meteo.csv")
feries_df = pd.read_csv("../data/raw/jours_feries_metropole.csv")

# Aper√ßu rapide
print(colis_df.head())
print(meteo_df.head())
print(feries_df.head())
```

```python
# Infos g√©n√©rales
print(colis_df.info())
print(meteo_df.info())
print(feries_df.info())
```

```python
print(colis_df.isnull().sum())
print(meteo_df.isnull().sum())
print(feries_df.isnull().sum())
```

```python
print(colis_df.describe())
print(meteo_df.describe())
```

```python
# Histogrammes
colis_df["nb_colis_entree"].hist(bins=30, color="steelblue", edgecolor="black")
plt.title("Distribution du nombre de colis entrants")
plt.xlabel("Nb colis")
plt.ylabel("Fr√©quence")
plt.show()

# Courbe m√©t√©o : temp√©rature max
plt.plot(meteo_df["date"], meteo_df["temperature_max"], color="orange")
plt.title("Temp√©rature maximale quotidienne √† Paris (2023-2024)")
plt.xlabel("Date")
plt.ylabel("Temp√©rature max (¬∞C)")

# Incline les labels de date pour √©viter le chevauchement
plt.xticks(rotation=45)

# Am√©liore la lisibilit√© (r√©duction du nombre de ticks affich√©s automatiquement)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # max 10 dates affich√©es

# Ajuste le layout pour √©viter la d√©coupe
plt.tight_layout()

plt.show()
```

üëâ Commentaires :

- Les variations saisonni√®res sont bien visibles (√©t√© plus chaud, hiver plus froid).

- Les dates sont inclin√©es pour une meilleure lisibilit√©.

```python
# Fusion par date
merged_df = colis_df.merge(meteo_df, on="date", how="left")

# Conversion de la colonne date
merged_df["date"] = pd.to_datetime(merged_df["date"])
```

üëâ Pourquoi :

- Pour pouvoir analyser l‚Äôimpact de la m√©t√©o sur le volume de colis.

- date au format datetime = utile pour les mod√®les temporels (future √©volution).

```python
# Calcul des corr√©lations uniquement sur les colonnes num√©riques
corr = merged_df.select_dtypes(include=["number"]).corr()

# Affichage de la heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corr√©lation")
plt.show()
```
